```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "sales_data.csv"  
sales_data = pd.read_csv(file_path)


print("Dataset successfully loaded.\n")
```

```python

print(sales_data.head())

print(f"Rows: {sales_data.shape[0]}, Columns: {sales_data.shape[1]}")

unique_products = sales_data["ProductName"].unique()
print(f"Unique Products: {unique_products}")

filtered_data = sales_data[sales_data["QuantitySold"] > 10]
print("\nFiltered DataFrame (QuantitySold > 10):")
print(filtered_data.head())

total_revenue = sales_data["Revenue"].sum()
print(f"\nTotal Revenue Generated: ${total_revenue:.2f}")

highest_revenue_row = sales_data.loc[sales_data["Revenue"].idxmax()]
highest_revenue_product = highest_revenue_row["ProductName"]
highest_revenue_value = highest_revenue_row["Revenue"]
print(f"Product with the Highest Revenue: {highest_revenue_product} (${highest_revenue_value:.2f})")

top_products = sales_data.groupby("ProductName")["QuantitySold"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top_products.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Top 5 Products by Quantity Sold", fontsize=14)
plt.xlabel("Product Name", fontsize=12)
plt.ylabel("Quantity Sold", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
